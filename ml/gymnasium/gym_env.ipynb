```python
import gymnasium as gym

import numpy as np
import mediapy as media
import matplotlib.pyplot as plt
```

```python
def show_video(frames, fps=30, **kwargs):
    media.show_video(frames, fps=fps, **kwargs)

def save_video(path, frames, fps=30, **kwargs):
    media.write_video(path, frames, fps=fps, **kwargs)
```

```python
gym.pprint_registry()
```

```python
# Make, interact, viz a Environment
env = gym.make('CartPole-v1', max_episode_steps=100, render_mode='rgb_array')
obs, info = env.reset()

episode_over = False
frames = []
while not episode_over:
    action = env.action_space.sample()
    obs, reward, terminate, truncate, info = env.step(action)
    frames.append(env.render())

    episode_over = truncate

env.close()
```

```python
show_video(frames, width=250, title='cart pole')
```
